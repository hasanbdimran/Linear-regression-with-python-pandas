```python
import pandas as pd
import urllib
import numpy as np
import os
import scipy.stats as sp
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
binpath = '/Users/ihasan/Downloads/Blogging/Linear regression with pandas/' 
fn_name = os.path.join(binpath,'data.xlsx')
df = pd.read_excel(fn_name)
df.set_index('Date', inplace=True)
```

```python
print (df.head())
df['OW2 As(mg/L)'].dropna().plot(marker='o', ls='');
```

```python
y=np.array(df['OW2 As(mg/L)'].dropna().values, dtype=float)
x=np.array(pd.to_datetime(df['OW2 As(mg/L)'].dropna()).index.values, dtype=float)
slope, intercept, r_value, p_value, std_err =sp.linregress(x,y)
xf = np.linspace(min(x),max(x),100)
xf1 = xf.copy()
xf1 = pd.to_datetime(xf1)
yf = (slope*xf)+intercept
```

```python
print('r = ', r_value, '\n', 'p = ', p_value, '\n', 's = ', std_err)
```

```python
f, ax = plt.subplots(1, 1)
ax.plot(xf1, yf,label='Linear fit', lw=3)
df['OW2 As(mg/L)'].dropna().plot(ax=ax,marker='o', ls='')
plt.ylabel('Arsenic concentration')
#ax.text(xf1[10], 0.2, infs)
ax.legend()
plt.savefig('/Users/ihasan/Downloads/Blogging/Linear regression with pandas/final_plot.png', dpi=300);
```
